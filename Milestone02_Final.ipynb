```python
#Step 1: Import random(shuffling), create global variables for Card Suit, Card Rank and Rank-Value dictionary
#Step 2: Create a card class that acts a prototype of every single card in the deck
#Step 3: Create a deck class that uses card class to generate and store the 52 cards
#Step 4: Create a hand class that represents the what the player/computer holds currently
#Step 5: Create a chip class to keep track of chips won/lost through bets
#Step 6: Use function to input bet from the user
#Step 7: Function for taking Hit
#Step 8: Function prompting player to either choose Hit or Stand
#Step 9: Displaying cards (both open for the players, one open and one closed for dealer)
#Step 10: Functions for the possible scenarios the game
#Step 11: THE SHOWDOWN
```

```python
#STEP 1:

import random
suit = ('Spades','Hearts','Diamonds','Clubs')
rank = ('Two','Three','Four','Five','Six','Seven','Eight','Nine','Ten','Jack','Queen','King','Ace')
value = {'Two':2,'Three':3,'Four':4,'Five':5,'Six':6,'Seven':7,'Eight':8,'Nine':9,'Ten':10,'Jack':10,'Queen':10,'King':10,'Ace':11}

playing = True
```

```python
#STEP 2:

class Card:
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank+" of "+self.suit
```

```python
#STEP 3:

class Deck:
    
    def __init__(self):
        self.deck = [] #Deck is empty at first
        for s in suit:
            for r in rank:
                self.deck.append(Card(s,r))
    
    def __str__(self):
        deck_has = ''
        card_count = 0
        for every_single_card in self.deck:
            deck_has += '\n'+every_single_card.__str__()
            card_count+=1
        return "The deck has: "+deck_has+'\nCount = '+str(card_count)
    
    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        single_card = self.deck.pop()  #Take the last card and deal it
        return single_card
```

```python
#STEP 4:

class Hand:
    
    def __init__(self):
        self.hand = []
        self.value = 0  
        self.count_ace = 0   
    
    def add_card(self,card):
        self.hand.append(card) 
        self.value += value[card.rank] 
        if card.rank == 'Ace':
            self.count_ace += 1
            
    def adjust_ace(self):
        if self.value > 21 and self.count_ace > 0: 
            self.value = self.value - 10      
            self.count_ace = self.count_ace - 1         
```

```python
#STEP 5:

class Chip:
    
    def __init__(self,total=100):
        self.total = total
        self.bet = 0
        
    def win_bet(self):
        self.total += self.bet
    
    def lose_bet(self):
        self.total -= self.bet
```

```python
#STEP 6:

def take_bet(chips):
    
    while True:
        try:
            chips.bet = int(input('How many chips would you like to bet? '))
        except ValueError:
            print('Sorry, a bet must be an integer!')
        else:
            if chips.bet > chips.total:
                print("Sorry, your bet can't exceed",chips.total)
            else:
                break
```

```python
#STEP 7:

def hit(deck,hand):
    
    hand.add_card(deck.deal())
    hand.adjust_ace()
```

```python
#STEP 8:

def hit_or_stand(deck,hand):
    global playing  
    
    while True:
        x = input("Would you like to Hit or Stand? Enter 'h' or 's' ")
        
        if x[0].lower() == 'h':
            hit(deck,hand)  

        elif x[0].lower() == 's':
            print("Player stands. Dealer is playing.")
            playing = False

        else:
            print("Sorry, please try again.")
            continue
        break   
```

```python
#STEP 9:

def show_some(player,dealer):
    print("\nDealer's Hand: ")
    print(" One of the cards is CLOSED")
    print('',dealer.hand[1])  
    print("\nPlayer's Hand: ")
    for card in player.hand:
        print(card)
    
def show_all(player,dealer):
    print("\nDealer's Hand: ")
    for card in dealer.hand:
        print(card)
    print("\nPlayer's Hand: ")
    for card in player.hand:
        print(card)
```

```python
#STEP 10:

def player_busts(player,dealer,chips):
    print("Player busts!")
    chips.lose_bet()

def player_wins(player,dealer,chips):
    print("Player wins!")
    chips.win_bet()

def dealer_busts(player,dealer,chips):
    print("Dealer busts!")
    chips.win_bet()
    
def dealer_wins(player,dealer,chips):
    print("Dealer wins!")
    chips.lose_bet()
    
def push(player,dealer):
    print("Dealer and Player tie! It's a push.")
```

```python
#STEP 11:

while True:
    
    #Print a welcome statement
    print("Hey there! Welcome to blackjack.")
    print("Fasten your seatbelts... The game is about to begin....!!")
    
    #Create a deck object, player object and add 2 cards to players, repeat for dealer
    
    deck = Deck()
    deck.shuffle()
    
    player_hand = Hand()
    player_hand.add_card(deck.deal())   #Card 1
    player_hand.add_card(deck.deal())   #Card 2
    
    dealer_hand = Hand()
    dealer_hand.add_card(deck.deal())
    dealer_hand.add_card(deck.deal())
    
    #Set up player's chips and get the bet input from player
    player_chips = Chip()
    take_bet(player_chips)
    
    #Show the cards of player and one from the dealer's
    
    show_some(player_hand,dealer_hand)
    
    while playing:    #To play game as long as player decides to stand
        
        #Check if the player wants to hit or stand
        hit_or_stand(deck,player_hand)
        
        #Show the cards (one of the dealer's card hidden)
        show_some(player_hand,dealer_hand)
        
        #Check if the player's card's value exceed 21 and call player_bust()
        if player_hand.value > 21:
            player_busts(player_hand,dealer_hand,player_chips)
            
            break
            
    # If the player isn't busted, play dealer's hand until its value exceeds that of player's hand, also player
    #decides to stand
    
    if player_hand.value <= 21:
        
        while dealer_hand.value < player_hand.value:
            hit(deck,dealer_hand)
            
        #Show all cards once dealer's hit ends
        show_all(player_hand,dealer_hand)
        
        #Check different winning scenarios
        if dealer_hand.value > 21:
            dealer_busts(player_hand,dealer_hand,player_chips)

        elif dealer_hand.value > player_hand.value:
            dealer_wins(player_hand,dealer_hand,player_chips)

        elif dealer_hand.value < player_hand.value:
            player_wins(player_hand,dealer_hand,player_chips)

        else:
            push(player_hand,dealer_hand)        
    
    # Inform Player of their chips total 
    print("\nPlayer's winnings stand at",player_chips.total)
    
    # Ask to play again
    new_game = input("Would you like to play another hand? Enter 'y' or 'n' ")
    
    if new_game[0].lower()=='y':
        playing=True
        continue
    else:
        print("Thanks for playing!")
        break
        
```
